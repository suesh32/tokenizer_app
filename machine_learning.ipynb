```python
import pandas as pd
import numpy as np

from sklearn.naive_bayes import MultinomialNB
```

# STEP3:学習,予測,結果

## 学習

```python
# 作成済みの学習データを読み込み
newlist = pd.read_csv("learn_data.csv",encoding="utf-8").fillna(0)
```

```python
# フラグ比率を確認
print(len(newlist[newlist["_flg_"]==0]))
print(len(newlist[newlist["_flg_"]==1]))
```

```python
# フラグ部分と項目部分に分割して変数に格納

## フラグ部分を格納
y=newlist["_flg_"]

## 項目部分を格納　学習に使わないカラムは削除
X=newlist.copy()
del X["_flg_"]
del X["_url_"]
del X["_date_"]
```

```python
# ベイズ学習器による学習　pythonのライブラリ関数を使用することで、１行で実施可能
# 作成したモデルをmnbという変数に格納
mnb = MultinomialNB().fit(X, y)
```

## 予測

```python
# 作成済のテストデータを読み込み
testlist = pd.read_csv("test_data.csv",encoding="utf-8").fillna(0)
```

```python
# 項目部分を格納　不要なカラムは削除
X2 = testlist.copy()

del X2["_url_"]
del X2["_date_"]
```

```python
# 作成したモデルにテストデータを適用　こちらも１行で実施可能
# 採用・不採用の予測結果をtmp_probという変数に格納
tmp_prob = mnb.predict_proba(X2)
```

## 出力

```python
# 形態素解析する前の学習データを読み込み
out_df = pd.read_csv("test_input.csv",encoding="utf-8")
```

```python
# 採用・不採用の予測結果をつけて、確信度が高い順にソートする
out_df["_prob_"] = tmp_prob[:,1]
out_df2 = out_df[["_url_","_title_","_prob_"]]
out_df2 = out_df2.sort_values("_prob_" , ascending=False)
```

```python
import urllib.request
from IPython.core.display import display
proxy_handler = urllib.request.ProxyHandler({
  'http': 'http://sueunix01:1693673795@kws.proxy.nic.fujitsu.com:8080/'})
opener = urllib.request.build_opener(proxy_handler) # 会社でやるとき用
#opener = urllib.request.build_opener() # 家でやるとき用
urllib.request.install_opener(opener)
data = urllib.request.urlopen('http://www.itmedia.co.jp/enterprise/')
display(data)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')

print(soup.prettify())
```

```python
a_tag_list = soup.find_all("a")
import datetime

article_list = []
for article in a_tag_list:
    url_and_title = [datetime.date.today(), article.get("href"), article.text]
    article_list.append(url_and_title)
```

```python
import pandas as pd
df = pd.DataFrame(article_list)
df.columns = ["_date_", "_url_", "_title_"]
df = df.dropna()
df
```

```python
df.to_csv("test_input.csv", index=False)
```

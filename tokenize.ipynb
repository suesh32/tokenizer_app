```python
import pandas as pd
import numpy as np
from janome.tokenizer import Tokenizer
```

```python
site_csv = pd.read_csv("group2.csv", encoding="cp932", parse_dates=["date"], infer_datetime_format=True)
# site_csv
```

```python
def get_noun(df):
    """Webページのタイトル文字列'title'を含むデータフレームを受け取り、
    各列に対して名詞だけ抽出し、名詞とその出現回数の辞書を作成する。
    そしてその辞書の配列を返す"""
    t = Tokenizer()
    # 最後に返す辞書の配列を用意
    word_vector_list = []
    
    for i,row in df.iterrows():
        list = []
        newdic = {}
        # 名詞をすべてlistに入れる(重複あり)
        for token in t.tokenize(str(row['title'])):
            hinshi = token.part_of_speech[:2]
            if(hinshi == '名詞'):
                list.append(token.surface)
        # 重複を排除した集合を作る
        word_set = set(list)
        # 集合の各要素に対して出現回数を調べ、辞書に入れる
        for word in word_set:
            newdic[word] = list.count(word)

        # できた辞書を辞書の配列へ追加
        word_vector_list.append(newdic)

#　　　　print(word_vector_list)
    return word_vector_list
```

```python
word_vector_list = get_noun(site_csv)
word_vector_pd = pd.DataFrame(word_vector_list)
# word_vector_pd.sort_index(axis=1) # 特にsort必要なかった
word_vector_pd
```

```python
newlist = pd.merge(site_csv, word_vector_pd, left_index=True, right_index=True)
del newlist['title']
newlist.fillna(0)
```

```python
newlist.to_csv("learn_data.csv", index=False)
```

ここからテストデータの作成

```python
test_input_csv = pd.read_csv("test_input.csv", encoding="utf-8", parse_dates=["date"], infer_datetime_format=True)
test_input_csv.dropna()
```

```python
# test_dataに出てくる名詞の数のデータフレームをまず作る
test_word_vector_list = get_noun(test_input_csv)
test_word_vector_pd = pd.DataFrame(test_word_vector_list)
# 教師データとテストデータの共通カラムのデータを抽出
test_word_vector_pd = test_word_vector_pd[np.intersect1d(word_vector_pd.columns, test_word_vector_pd.columns)]
# ダミーのデータフレームを作成　教師データの全カラムを持ち、値が全0の1レコードのみ。
tmp_DF=pd.DataFrame(np.repeat(0,len(word_vector_pd.columns))).T
tmp_DF.columns=word_vector_pd.columns
# ダミーデータフレームのカラム部分とテストデータと結合して、欠損を0で埋める。
test_word_vector_pd = pd.concat([tmp_DF[1:],test_word_vector_pd]).fillna(0)
# test_word_vector_pd2.sort_index(axis=1) # 特にsort必要なかった
test_word_vector_pd
```

```python
testlist = pd.merge(test_input_csv, test_word_vector_pd, left_index=True, right_index=True)
del testlist['title']
testlist.fillna(0)
```

```python
# testlist.to_csv("test_data.csv", index=False, date_format="%Y/%m/%d")
testlist.to_csv("test_data.csv", index=False)
```

```python
import pandas as pd
from janome.tokenizer import Tokenizer
```

```python
site_csv = pd.read_csv("採用不採用リスト.csv", encoding="cp932", parse_dates=["date"], infer_datetime_format=True)
site_csv
```

```python
t = Tokenizer()
word_vector_list = []
for i,row in site_csv.iterrows():
    list = []
    newdic = {}
    for token in t.tokenize(row['title']):
        hinshi = token.part_of_speech[:2]
        if(hinshi == '名詞'):
            list.append(token.surface)
    word_set = set(list)
    for word in word_set:
        newdic[word] = list.count(word)
    print("===============================")
    print(newdic)
    word_vector_list.append(newdic)
print("-------------------------------")
print(word_vector_list)
```

```python
word_vector_pd = pd.DataFrame(word_vector_list)
word_vector_pd
```

```python
newlist = pd.merge(site_csv, word_vector_pd, left_index=True, right_index=True)
newlist
```

```python
del newlist['title']
newlist.fillna(0)
```

```python
newlist.to_csv("learn_data.csv", index=False)
```
